##### (5) 분석 결과 도출

가공된 데이터를 기반으로 분석을 진행했습니다.<br>
1. 평균 킬 수 계산<br>
2. 평균 생존 시간 계산<br>
3. 평균 순위 계산<br>

````python
import pandas as pd

class PlayerStats:
    def __init__(self, name):
        self.name = name
        self.total_kills = 0
        self.total_timeSurvived = 0
        self.total_rank = 0
        self.match_count = 0

    def add_match(self, kills, timeSurvived, rank):
        self.total_kills += int(kills)
        self.total_timeSurvived += int(timeSurvived)
        self.total_rank += int(rank)
        self.match_count += 1

    def get_averages(self):
        if self.match_count == 0:
            return 0, 0, 0
        avg_kills = self.total_kills / self.match_count
        avg_timeSurvived = self.total_timeSurvived / self.match_count
        avg_rank = self.total_rank / self.match_count
        return avg_kills, avg_timeSurvived, avg_rank

def analyze_player_stats(file_path):
    df = pd.read_csv(file_path)

    players = {}

    for _, row in df.iterrows():
        name = row["name"]
        kills = row["kills"]
        timeSurvived = row["timeSurvived"]
        rank = row["rank"]

        if name not in players:
            players[name] = PlayerStats(name)

        players[name].add_match(kills, timeSurvived, rank)

    for name, stats in players.items():
        avg_kills, avg_timeSurvived, avg_rank = stats.get_averages()
        print(f"{name}: 평균 킬 {round(avg_kills, 2)}, 평균 생존시간 {round(avg_timeSurvived, 2)}초, 평균 순위 {round(avg_rank, 2)}")

analyze_player_stats("player_stats.csv")
````

##### (10) 데이터를 분석하기 위해 가공한 코드 원본

API KEY를 통해 획득한 데이터가 저장된 'player_stats.csv'에서 <br>
데이터를 분석해서 가공하는데 사용한 코드 원본이다.

````python
import pandas as pd

df = pd.read_csv("player_stats.csv")

player_name = []

name = list(df["name"])
kill = list(df["kills"])
timeSurvived = list(df["timeSurvived"])
rank = list(df["rank"])


for n in name:
    if n not in player_name:
        player_name.append(n)

sum_kills = {}
sum_timeSurvived = {}
sum_rank = {}
number = {}

for i in range(len(name)):
    n = name[i]
    if n not in sum_kills:
        sum_kills[n] = 0
        sum_timeSurvived[n] = 0
        sum_rank[n] = 0
        number[n] = 0

    sum_kills[n] += int(kill[i])
    sum_timeSurvived[n] += int(timeSurvived[i])
    sum_rank[n] += int(rank[i])
    number[n] += 1

for name in player_name:
    avg_kill = sum_kills[name] / number[name]
    avg_timeSurvived = sum_timeSurvived[name] / number[name]
    avg_rank = sum_rank[name] / number[name]

    print(f"{name}:  평균 킬 {round(avg_kill,2)}, 평균 생존시간 {round(avg_timeSurvived,2)}초, 평균 순위 {round(avg_rank,2)}")
````
